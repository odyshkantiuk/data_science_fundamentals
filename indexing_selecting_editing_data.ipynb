## Мітко‐орієнтований вибір в бібліотеці Pandas

```python
# Вивід стовпця заданого по імені по індексу елемента за допомогою мітко‐орієнтований вибір в бібліотеці Pandas (loc)
covid_data.loc[0, 'iso_code']
```

```python
# Вивід підмножини стовпців, імена яких задані списком за допомогою мітко‐орієнтований вибір в бібліотеці Pandas (loc)
covid_data.loc[:, ['location','date', 'total_cases']]
```

```python
# Вивід перших 3 рядків з підмножини стовпців, імена яких задані списком за допомогою мітко‐орієнтований вибір в бібліотеці Pandas (loc)
covid_data.loc[:2 , ['location','date', 'total_cases']]
```

```python
# Вивід рядків з кроком у заданому діапазоні з підмножини стовпців, імена яких задані списком за допомогою мітко‐орієнтований вибір в бібліотеці Pandas (loc)
covid_data.loc[1:100:14 , ['location','date', 'total_cases']]
```

## Умовний вибір

```python
# Вивід даних, де кількість лікарняних ліжок на тисячу осіб більша за середню
hospital_beds_per_thousand_mean=covid_data.hospital_beds_per_thousand.mean()
covid_data.loc[covid_data.hospital_beds_per_thousand>hospital_beds_per_thousand_mean]
```

```python
# Вивід даних, де кількість лікарняних ліжок на тисячу осіб більша за середню і кількість загальна захворювань менша за середню
total_cases_mean=covid_data.total_cases.mean()
covid_data.loc[(covid_data.hospital_beds_per_thousand>hospital_beds_per_thousand_mean) & (covid_data.total_cases<total_cases_mean)]
```

```python
# Вивід даних по Україні за допомогою «is in» який дозволяє вибирати дані, значення яких є у списку значень
covid_data.loc[covid_data.location.isin(['Ukraine'])]
```

```python
# Вивід даних за 2020-02-24 та за 2021-02-24 для порівняння за допомогою «is in» який дозволяє вибирати дані, значення яких є у списку значень
covid_data.loc[covid_data.date.isin(['2020-02-24', '2021-02-24'])]
```

```python
# Вивід порожніх даних по стовпцю «total_cases» за допомогою «isnull» який дозволяє виділити значення, які є порожніми (NaN)
covid_data.loc[covid_data.total_cases.isnull()]
```
